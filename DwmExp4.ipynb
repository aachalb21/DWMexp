```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
arr = np.array([1, 3, 6, 8, 12, 18, 25, 28, 35, 40])
```

```python
k1 = arr[0]
k2 = arr[1]
print(k1,k2)
```

```python
while True:
  k1new = []
  k2new = []
  for i in arr:
    if abs(k1 - i) <= abs(k2 - i):
      k1new.append(i)
    else:
      k2new.append(i)
  print(k1new)
  print(k2new)
  m1 = np.mean(k1new)
  m2 = np.mean(k2new)
  print("The centroids are: ",m1,m2)
  print()
  if k1 == m1 and k2 == m2:
   break
  k1 = m1
  k2 = m2
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(k1new, [0]*len(k1new), color='blue', label='Cluster 1')
plt.scatter(k2new, [0]*len(k2new), color='red', label='Cluster 2')
plt.scatter(m1, 0, marker='x', s=100, color='blue', label='Centroid 1')
plt.scatter(m2, 0, marker='x', s=100, color='red', label='Centroid 2')
plt.xlabel('Data Points')
plt.ylabel('')
plt.title('K-Means Clustering Result')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = np.array([[1, 2], [1.5, 1.8], [5, 8], [8, 8], [1, 0.6], [9, 11]])
```

```python
k = int(input("Enter the number of clusters (k): "))
```

```python
# Perform k-means clustering
kmeans = KMeans(n_clusters=k, random_state=0).fit(data)
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
print("Cluster Labels:", labels)
print("Centroids:", centroids)
```

```python
plt.figure(figsize=(8, 6))
for i in range(k):
plt.scatter(data[labels == i, 0], data[labels == i, 1], label=f"Cluster {i + 1}")
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=100, c='red', label='Centroids')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("2D K-Means Clustering")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
for i in range(k):
    # Indent the code block within the for loop
    plt.scatter(data[labels == i, 0], data[labels == i, 1], label=f"Cluster {i + 1}")

plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=100, c='red', label='Centroids')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("2D K-Means Clustering")
plt.legend()
plt.grid(True)
plt.show()
```
